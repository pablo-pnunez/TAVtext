```python
from src.experiments.Common import load_best_model
import pandas as pd
```

```python
best_model = pd.read_csv("models/best_models.csv")
all_data = []
for _, row in best_model.iterrows():
    model = row["model"]
    dataset = row["dataset"]
    subset = row["subset"]
    model_md5 = row["model_md5"]

    save_path = f'models/{model}/{dataset}/{subset}/{model_md5}/log.csv'
    model_data = pd.read_csv(save_path)

    model_train_time = model_data["e_time"].sum()

    model_class = load_best_model(model=model, dataset=dataset, subset=subset)
    model_params = model_class.MODEL.count_params()

    all_data.append((model, dataset, subset, model_md5, model_train_time, model_params))

pd.DataFrame(all_data, columns=["model","dataset","subset","model_md5","model_train_time","model_params"]).to_excel("model_stats.xlsx")
```
