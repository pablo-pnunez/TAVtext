```python
from transformers import pipeline

# Cargamos el modelo para chat en español (BETO)
model_name = "dccuchile/bert-base-spanish-wwm-uncased"
chatbot = pipeline("conversational", model=model_name)

# Conversación con el chatbot
conversation = [
    {"role": "user", "content": "Hola, ¿cómo estás?"},   # Mensaje del usuario
    {"role": "assistant", "content": "Hola, soy un chatbot en español. ¿En qué puedo ayudarte?"},  # Respuesta del chatbot
    {"role": "user", "content": "¿Cuál es el clima hoy?"},   # Mensaje del usuario
    {"role": "assistant", "content": "Lo siento, no tengo acceso a información en tiempo real. Pero puedo ayudarte en otras cosas."},  # Respuesta del chatbot
]

for step in conversation:
    user_input = step["content"]
    print(f"Usuario: {user_input}")
    response = chatbot(user_input)
    print(f"Chatbot: {response['generated']['text']}\n")
```
